```python
import numpy as np
from matplotlib import pyplot as plt
# %matplotlib notebook
%matplotlib inline
from skimage.io import imread,imshow,imsave
```

```python
im = imread('walking.jpg')
plt.figure()
plt.gray()
plt.imshow(im)
plt.show()
```

```python
im_vector=im.reshape([1,im.size])
f=open("imvector.txt","w")
for i in im_vector[0]:
    f.write(f"{i} ")
```

```python
from scipy.signal import convolve2d
from skimage.filters import gaussian

gy = np.array([[-1.,-2.,-1.],[0.,0.,0.],[1.,2.,1.]])
gx = gy.T # transpose operator

print(gy)
print(gx)
```

```python
Gy = convolve2d(im, gy)
Gx = convolve2d(im, gx)

plt.figure()
plt.subplot(1,2,1)
plt.imshow(np.abs(Gy))
plt.subplot(1,2,2)
plt.imshow(np.abs(Gx))
plt.show()
```

```python
Gx_vector=Gx.reshape([1,Gx.size])
fGx=open("gxvector.txt","w")
for i in Gx_vector[0]:
    fGx.write(f"{i} ")
```

```python
Gy_vector=Gx.reshape([1,Gy.size])
fGy=open("gyvector.txt","w")
for i in Gy_vector[0]:
    fGy.write(f"{i} ")
```

```python
s = np.sqrt(Gy**2+Gx**2)
d = np.arctan2(Gy,Gx)

plt.figure()
plt.subplot(1,2,1)
plt.imshow(s)
plt.subplot(1,2,2)
plt.imshow(d, cmap=plt.cm.viridis)
plt.colorbar()
plt.show()
```
